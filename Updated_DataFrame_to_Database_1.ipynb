```python
import pandas as pd
import glob

# Assuming all your Excel files are in the same directory
file_paths = [
    "C:\\Users\\Green Gen Tech\\Downloads\\apsrtc_redbus_project_redbus_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\ksrtc_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Documents\\kadamba_redbus_project_redbus_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\rsrtc_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\tsrtc_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\south_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\hrtc_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\up_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\chandigarh_redbus_project_redbus_new_1.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\wbtc_redbus_project_redbus_new_1.xlsx"
]

# Creating a list of DataFrames from each file
dataframes = [pd.read_excel(file) for file in file_paths]

# Combining DataFrames
combined_df = pd.concat(dataframes)

# Resetting the index
combined_df = combined_df.reset_index(drop=True)

# Printing the combined DataFrame
combined_df
```

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib.parse

host = "localhost"
user = "root"
password = "Nagunaveen@143"  
database = "bus"
password_encoded = urllib.parse.quote_plus(password)
connection_string = f"mysql+mysqlconnector://{user}:{password_encoded}@{host}/{database}"
engine = create_engine(connection_string)
csv_file = r"C:\\Users\\Green Gen Tech\\Downloads\\combined_redbus_data_final.xlsx"
df = pd.read_csv(csv_file)
table_name = 'redbus'
try:
    df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
    print(f"Data successfully written to '{table_name}' table in MySQL.")
except Exception as e:
    print(f"Error writing data to MySQL: {e}")
engine.dispose()
```
