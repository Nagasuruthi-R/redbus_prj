***I created xpath for combined dataframes to create database***

```python
import pandas as pd

# List of file paths
file_paths = [
    "C:\\Users\\Green Gen Tech\\Downloads\\andhrapradesh_redbus_project_redbus.xls.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\kerala_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\kadamba_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\rajasthan_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\telangana_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\south_bengal_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\haryana_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\uttarpradesh_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\chandigarh_redbus_project_redbus.xlsx",
    "C:\\Users\\Green Gen Tech\\Downloads\\west_bengal_redbus_project_redbus.xlsx"
]

# Creating a list of DataFrames from each file
dataframes = [pd.read_excel(file) for file in file_paths]

# Combining DataFrames
combined_df = pd.concat(dataframes)

# Resetting the index
combined_df = combined_df.reset_index(drop=True)

# Saving the combined DataFrame to an Excel file
output_file_path = "C:\\Users\\Green Gen Tech\\Downloads\\combined_redbus_data.xlsx"
combined_df.to_excel(output_file_path, index=False)

# Printing a confirmation message
print(f"Combined DataFrame saved to {output_file_path}")
```

***Added those datas into the table naming "redbus_routes" and connected every datas in mysql by using sqlalchemy and create_engine***

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib.parse

host = "localhost"
user = "root"
password = "Nagunaveen@143"  
database = "bus_informations"
password_encoded = urllib.parse.quote_plus(password)
connection_string = f"mysql+mysqlconnector://{user}:{password_encoded}@{host}/{database}"
engine = create_engine(connection_string)
excel_file = r"C:\Users\Green Gen Tech\Downloads\combined_redbus_data.xlsx"
df = pd.read_excel(excel_file)
table_name = 'redbus_routes'
try:
    df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
    print(f"Data successfully written to '{table_name}' table in MySQL.")
except Exception as e:
    print(f"Error writing data to MySQL: {e}")
engine.dispose()
```
